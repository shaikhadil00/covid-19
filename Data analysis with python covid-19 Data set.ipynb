```python
statewise=pd.pivot_table(df, values = ["Total Cases" , "Deaths" , "Discharged"],index = "State/UTs" , aggfunc = max)
```

```python
statewise["Recovery Rate"] = statewise["Discharged"]*100/statewise["Total Cases"] 
```

```python
statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Total Cases"] 
```

```python
statewise = statewise.sort_values(by = "Total Cases", ascending = False)
```

```python
#Top 10 active cases states

top_10_Active_cases = df.groupby(by = 'State/UTs').max()[['Active','Population']].sort_values(by = ['Active'], ascending = False).reset_index()
```

```python
fig = plt.figure(figsize=(16,9))
```

```python
plt.title("Top 10 states with most Active cases in india" , size = 15)
```

```python
ax = sns.barplot(data = top_10_Active_cases.iloc[:10], y = "Active" , x = "State/UTs" , linewidth = 2 , edgecolor = 'blue')
```

# They are now overlaping with each other so we have check with 'x' and 'y' label with adding of #Top 10 active cases states

```python
#Top 10 active cases states

top_10_Active_cases = df.groupby(by = 'State/UTs').max()[['Active','Population']].sort_values(by = ['Active'], ascending = False).reset_index()

fig = plt.figure(figsize=(16,9))

plt.title("Top 10 states with most Active cases in india" , size = 15)

ax = sns.barplot(data = top_10_Active_cases.iloc[:10], y = "Active" , x = "State/UTs" , linewidth = 2 , edgecolor = 'black')

plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
#Top states with Highest Deaths

top_10_Deaths = df.groupby(by = 'State/UTs').max()[['Deaths','Population']].sort_values(by = ['Deaths'], ascending = False).reset_index()

fig = plt.figure(figsize=(16,9))

plt.title("Top states with Highest Deaths in india" , size = 15)

ax = sns.barplot(data = top_10_Deaths.iloc[:10], y = "Deaths" , x = "State/UTs" , linewidth = 2 , edgecolor = 'black')

plt.xlabel("States")
plt.ylabel("Total Deaths")
plt.show()
```
